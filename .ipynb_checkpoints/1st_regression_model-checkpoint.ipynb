```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
exp_runs = pd.read_excel('Experimental Runs.xlsx', index_col = 0)
```

```python
exp_runs
```

```python
one_min_runs = pd.read_excel('Experimental_Runs(Reg_data).xlsx', index_col = 0)
```

```python
one_min_runs
```

```python
df = pd.DataFrame(one_min_runs)
```

```python
df[1:3]
```

```python
y = df[['RMS']]
X = df[['Cutting Speed (m/min)', 'Feed (mm/rec)', 'Depth of cut (mm)', 'RPM', 'Diameter(mm)']]
y = y[1:]
X = X[1:]
```

```python
y[9:10] = 0
y.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 5)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_test = np.array(y_test)
y_test
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
r_2_score = r2_score(y_test, y_pred)
r_2_score
```

```python
explained_variance_score(y_test, y_pred)
```

```python
 max_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
 mean_squared_error(y_test, y_pred)
```
